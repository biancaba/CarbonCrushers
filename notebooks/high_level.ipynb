```python
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt

sns.set_palette("husl", 8)
None
```

```python
df = pd.read_csv("Food_Production.csv")
```

```python
df
```

```python
filt_df = df[["Food product", "Animal Feed"]]
```

```python
filt_df["Production"] = df[["Farm", "Processing"]].sum(axis=1)
filt_df["Distribution"] = df[["Transport", "Packging"]].sum(axis=1)
filt_df
```

### Adding in score columns

```python
af_df = filt_df[filt_df["Animal Feed"] != 0]
```

```python
thresh_1 = af_df["Animal Feed"].quantile(0.2)
thresh_2 = af_df["Animal Feed"].quantile(0.4)
thresh_3 = af_df["Animal Feed"].quantile(0.6)
thresh_4 = af_df["Animal Feed"].quantile(0.8)

thresh_1, thresh_2, thresh_3, thresh_4
```

```python
filt_df["Animal_Feed_Score"] = pd.Series([0 for x in range(len(filt_df))])
for i in range(len(filt_df)):
    val = filt_df["Animal Feed"].iloc[i]
    if val == 0:
        pass
    elif val < thresh_1:
        filt_df["Animal_Feed_Score"].iloc[i] = 1
    elif val < thresh_2:
        filt_df["Animal_Feed_Score"].iloc[i] = 2
    elif val < thresh_3:
        filt_df["Animal_Feed_Score"].iloc[i] = 3
    elif val < thresh_4:
        filt_df["Animal_Feed_Score"].iloc[i] = 4
    else:
        filt_df["Animal_Feed_Score"].iloc[i] = 5
```

```python
thresh_1 = filt_df["Production"].quantile(0.2)
thresh_2 = filt_df["Production"].quantile(0.4)
thresh_3 = filt_df["Production"].quantile(0.6)
thresh_4 = filt_df["Production"].quantile(0.8)

thresh_1, thresh_2, thresh_3, thresh_4
```

```python
filt_df["Production_Score"] = pd.Series([0 for x in range(len(filt_df))])
for i in range(len(filt_df)):
    val = filt_df["Production"].iloc[i]
    if val < thresh_1:
        filt_df["Production_Score"].iloc[i] = 1
    elif val < thresh_2:
        filt_df["Production_Score"].iloc[i] = 2
    elif val < thresh_3:
        filt_df["Production_Score"].iloc[i] = 3
    elif val < thresh_4:
        filt_df["Production_Score"].iloc[i] = 4
    else:
        filt_df["Production_Score"].iloc[i] = 5
```

```python
thresh_1 = filt_df["Distribution"].quantile(0.2)
thresh_2 = filt_df["Distribution"].quantile(0.4)
thresh_3 = filt_df["Distribution"].quantile(0.6)
thresh_4 = filt_df["Distribution"].quantile(0.8)

thresh_1, thresh_2, thresh_3, thresh_4
```

```python
filt_df["Distribution_Score"] = pd.Series([0 for x in range(len(filt_df))])
for i in range(len(filt_df)):
    val = filt_df["Distribution"].iloc[i]
    if val < thresh_1:
        filt_df["Distribution_Score"].iloc[i] = 1
    elif val < thresh_2:
        filt_df["Distribution_Score"].iloc[i] = 2
    elif val < thresh_3:
        filt_df["Distribution_Score"].iloc[i] = 3
    elif val < thresh_4:
        filt_df["Distribution_Score"].iloc[i] = 4
    else:
        filt_df["Distribution_Score"].iloc[i] = 5
```
